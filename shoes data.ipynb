```python
def Brand (x):
    if x>=325 and x<=300000:
        return 'low_b'
    elif x>300000 and x<=600000:
        return 'Mid_b'
    elif x>600000 and x<=900000:
        return 'high_b'
    else:
        return 'super_b'
```
